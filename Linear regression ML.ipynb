## Preparation for  Model Building

```python
## dividing dataset into dependent and inependent features
```

```python
x= df.iloc[:,:-1]
```

```python
x = x.drop(columns=['price'])
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
x
```

```python
## train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=.3 , random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
## standard scaling
# while using gradient descent main aim is to come near to global min
# we have to standardised features as all of them have different i units 
# to reach to global min we have to make sure all our independent features units must have same scale
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
# do not use fit so that model do not know much about test data
x_test= scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
## print coefficients
```

```python
reg.coef_
```

```python
reg.intercept_
#
```

```python
## on which parameter model has been trained
reg.get_params()
```

```python
## prediction with test data
reg_pred= reg.predict(x_test)
```

```python
reg_pred
```

```python
# plot a scatter plot for prediction
```

```python
plt.scatter(y_test,reg_pred)
# reg_pred is true value 
# if prediction loks linear it is good
```

```python
# residula is error
residual = y_test - reg_pred
```

```python
residual
```

```python
#plot this residuals
```

```python
sns.displot(residual , kind='kde')
```

```python
## scatter plot wrt to predictions and residuals
plt.scatter(reg_pred , residual)

# these are scatter and uniform distributed wrt to test data and residual
```

```python
## performance matrices
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
print(mean_absolute_error(y_test , reg_pred))
print(mean_squared_error(y_test , reg_pred))
print(np.sqrt(mean_squared_error(y_test , reg_pred)))
```

# r squared and adjusted r squared

```python
# adjusted r2 is less than r2
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test , reg_pred)
print(score)
```
